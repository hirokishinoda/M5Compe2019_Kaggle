各店舗，状態ごとの値段の遷移を見ていく

```python
plt.figure()
for i in range(10):
    plt.plot(train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 6:].values,
             label=train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 5]);
plt.title('HOBBIES_1_002 sales')
plt.legend();
```

```python
plt.figure(figsize=(12, 4))
for i in range(10):
    plt.plot(train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 6:].rolling(30).mean().values,
             label=train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 5]);
plt.title('HOBBIES_1_002 sales, rolling mean 30 days')
plt.legend();

plt.figure(figsize=(12, 4))
for i in range(10):
    plt.plot(train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 6:].rolling(60).mean().values,
             label=train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 5]);
plt.title('HOBBIES_1_002 sales, rolling mean 60 days')
plt.legend();

plt.figure(figsize=(12, 4))
for i in range(10):
    plt.plot(train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 6:].rolling(90).mean().values,
             label=train_df.loc[train_df['item_id'] == 'HOBBIES_1_002'].iloc[i, 5]);
plt.title('HOBBIES_1_002 sales, rolling mean 90 days')
plt.legend();
```

### カテゴリIDと部門IDを表にして表示する

```python
ca_1_sales = train_df.loc[train_df['store_id'] == 'CA_1']
pd.crosstab(ca_1_sales['cat_id'], ca_1_sales['dept_id'])
```

食品が3部門，おもちゃと家具？が2部門

```python
plt.figure(figsize=(12, 4))
for d in ca_1_sales['dept_id'].unique():
    store_sales = ca_1_sales.loc[ca_1_sales['dept_id'] == d]
    store_sales.iloc[:, 6:].sum().rolling(30).mean().plot(label=d)
plt.title('CA_1 sales by department, rolling mean 30 days')
plt.legend(loc=(1.0, 0.5));
```

他に比べてFOOD_3が高いですね

## 一つの部門について見ていく

```python
train_df.loc[train_df['dept_id'] == 'HOBBIES_1']
```

```python
train_df.loc[train_df['dept_id'] == 'HOBBIES_1', 'item_id'].nunique()
```

uniqueなアイテムは416個

```python
hobbies_1_df = train_df.loc[train_df['dept_id'] == 'HOBBIES_1']
plt.figure(figsize=(12, 6))
for d in hobbies_1_df['store_id'].unique():
    store_df = hobbies_1_df.loc[hobbies_1_df['store_id'] == d]
    store_df.iloc[:, 6:].sum().rolling(30).mean().plot(label=d)
plt.title('HOBBIES_1 df by stores, rolling mean 30 days')
plt.legend(loc=(1.0, 0.5));
```

CA_1とCA_2が高いですね

```python
hobbies_1_prices_df = sell_prices_df.loc[sell_prices_df['item_id'].str.contains('HOBBIES_1')]
plt.figure(figsize=(12, 6))
for d in hobbies_1_prices_df['store_id'].unique():
    small_df = hobbies_1_prices_df.loc[hobbies_1_prices_df['store_id'] == d]
    grouped = small_df.groupby(['wm_yr_wk'])['sell_price'].mean()
    plt.plot(grouped.index, grouped.values, label=d)
plt.legend(loc=(1.0, 0.5))
plt.title('HOBBIES_1 mean sell prices_df by store');
```

少しづつ値段が上昇している
